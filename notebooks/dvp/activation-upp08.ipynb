```python
import numpy as np
import matplotlib.pyplot as plt 

from collections import defaultdict
from mckit import read_mcnp
from pathlib import Path
from r2s_rfda.fetch import load_result_config, load_data, load_result_config
from r2s_rfda import utils
from r2s_rfda.launcher import load_config

from mckit import *

%matplotlib inline
```

```python
path = Path(r'd:\dvp\dev\mcnp\upp\wrk\up08\r2s')
config = load_config(path)
result_conf = load_result_config(path)
```

```python
time = utils.convert_time_literal('12d')
time_labels = list(sorted(result_conf['gamma'].keys()))
time += time_labels[config['zero']]
closest_lab = utils.find_closest(time, time_labels)
closest_lab
```

```python
data_paths = result_conf
act_paths = data_paths['activity']
```

```python
act_paths
```

```python
act_data = load_data(path / act_paths[closest_lab])
```

```python
cell_activity = defaultdict(float)
for cnt, ((g, c, i, j, k), act) in enumerate(act_data.iter_nonzero()):
    if cnt % 10000 == 0:
        print(cnt)
    cell_activity[c] += act
    #print(act_data.gbins[g], c, i, j, k, act)
    #i += 1
    #if i == 1000:
    #    break
```

```python
cell_activity
```

```python
#config['masses']
cell_mass = defaultdict(float)
for (c, i, j, k), ms in config['masses'].items():
    cell_mass[c] += ms
```

```python
cell_mass
```

```python
cell_isot_activity = defaultdict(lambda: defaultdict(float))
for cnt, ((g, c, i, j, k), act) in enumerate(act_data.iter_nonzero()):
    isotope = act_data.gbins[g]
    cell_isot_activity[c][isotope] += act
    #print(act_data.gbins[g], c, i, j, k, act)
    #i += 1
    #if i == 1000:
    #    break
```
